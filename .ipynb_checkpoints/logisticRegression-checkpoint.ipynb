```python
import numpy as np
import pylab as pl
from math import exp, log
import scipy.optimize as sp
import matplotlib.pyplot as plt
# from scipy.special import expit #Vectorized sigmoid function


# Plot Data

def plotData():
    
        plt.figure(figsize=(10,6))
        plt.plot(pos[:,1],pos[:,2],'k+',label='Admitted')
        plt.plot(neg[:,1],neg[:,2],'yo',label='Not admitted')
        plt.xlabel('Exam 1 score')
        plt.ylabel('Exam 2 score')
        plt.legend()
        plt.grid(True)

        boundary_xs = np.array([np.min(X[:,1]), np.max(X[:,1])])
        boundary_ys = (-1./thetaFinal[2])*(thetaFinal[0] + thetaFinal[1]*boundary_xs)
        plt.plot(boundary_xs,boundary_ys,'b-',label='Decision Boundary')
        plt.legend()
        plt.show()
    

# Sigmoid Function

def sigmoid (x):
    
    result = []
    size = np.shape(x)
    
    try:
        for i in x.flat:
            result.append((1 / (1 + exp(-i))))
        return np.reshape(result, size)
    except:
        return (1 / (1 + exp(-x)))


# Cost Function and Gradient

def computeCost(theta, X, y, lmbda):

    prediction = np.dot(X, theta)
    hypothesis = sigmoid(prediction)
    J = (1/m) * np.sum( (np.dot(-np.transpose(y), np.log(hypothesis))) - np.dot(np.transpose(1 - y), np.log(1 - hypothesis)))
    regTerm = (lmbda / (2*m)) * np.sum(np.dot(np.transpose(theta[1:]), theta[1:])) # Skip theta0
    return J + regTerm

def hypothesis(theta, X): 
    return sigmoid(np.dot(X, theta))

# def computeCost(theta, X, y, lmbda): 
    
#     (m, n) = np.shape(X)
#     termOne = np.dot(- np.transpose(np.array(y)), np.log(hypothesis(theta, X)))
#     termTwo = np.dot(np.transpose((1 - np.array(y))), np.log(1 - hypothesis(theta, X)))
#     regTerm = (lmbda / 2) * np.sum(np.dot(np.transpose(theta[1:]), theta[1:])) #Skip theta0
#     return float( (1/m) * ( np.sum(termOne - termTwo) + regTerm ) )


def computeGradient(X, y, theta, lmbda):
    
    (m, n) = np.shape(X)

    prediction = np.dot(X, theta) # (m x n) * (n x 1)
    
    hypothesis = sigmoid(prediction) # (m x 1)
    
    errorVector = hypothesis - y # (m x 1)
    
    grad = (1/m) * np.dot(np.transpose(X), errorVector) # (n x m) * (m x 1)
    
    theta[0] = 0
    regTermGrad = (lmbda/m) * theta
    
    return grad + regTermGrad

def mapFeature(x1col, x2col):
    
#     Function that takes in a column of n- x1's, a column of n- x2s, and builds
#     a n- x 28-dim matrix of featuers as described in the homework assignment

    degrees = 6
    out = np.ones( (x1col.shape[0], 1) )

    for i in range(1, degrees+1):
        for j in range(0, i + 1):
            term1 = x1col ** (i - j)
            term2 = x2col ** (j)
            term  = (term1 * term2).reshape(term1.shape[0], 1) 
            out   = np.hstack((out, term))
    return out

def predict(theta, X):
    
    prediction = np.dot(X, theta)
    p = np.zeros(np.shape(prediction))
    for i in range(np.shape(X)[0]):
        if sigmoid(prediction[i]) < 0.5:
            p[i] = 0
        else:
            p[i] = 1
            
    return p
```

```python
##########################################################################################
# Ex2Data1
# No Regularisation
##########################################################################################


data = np.loadtxt('/Users/Zohair/Library/Mobile Documents/com~apple~CloudDocs/Education/Machine Learning/machine-learning-ex2/ex2/ex2data1.txt', delimiter = ",")

# Run Logistic Regression

(m, n) = np.shape(data)

# Add Column Of Ones
temp = np.ones((m, 1)) 
data = np.hstack((temp, data))

(m, n) = np.shape(data)

# Get X and y from data
(X, y) = np.hsplit(data, (n - 1,))

#Divide the sample into two: ones with positive classification, one with null classification
pos = np.array([X[i] for i in range(np.shape(X)[0]) if y[i] == 1])
neg = np.array([X[i] for i in range(np.shape(X)[0]) if y[i] == 0])

# Initialise Theta
initialTheta = np.zeros((n - 1, 1))

# Paramaters

lmbda = 0

# Initial Cost And Gradient
cost = computeCost(initialTheta, X, y, lmbda)
grad = computeGradient(X, y, initialTheta, lmbda)

print ("Cost at initial theta (zeros)", cost)
# print ("Gradient at initial theta (zeros):")
# print (np.shape(grad)) Correct

# def testFunction(theta):
#      return computeCost(theta, X, y, lmbda)

# def testFunctionTwo(theta):
#     return computeGradient(X, y, theta, lmbda)

# Minimise Cost

# y must be 1-D for minimisation functions to work
y = y.flatten()

thetaFinal = sp.fmin(computeCost, x0 = initialTheta, args = (X, y, lmbda))
# thetaFinal = sp.fmin_cg(testFunction, x0=initialTheta, fprime=testFunctionTwo)

print(thetaFinal)

testCase = np.array([[1, 45, 85]])
prob = sigmoid (np.dot(testCase, thetaFinal))
print(prob)


p = predict(thetaFinal, X)
print ("Train Accuracy:", np.mean(p == y) * 100, "%")

# Plot Data
plotData()

```

```python
##########################################################################################
# Ex2Data2
# With Regularisation
##########################################################################################



def plotData():
    
        plt.figure(figsize=(10,6))
        plt.plot(pos[:,1],pos[:,2],'k+',label='y = 1')
        plt.plot(neg[:,1],neg[:,2],'yo',label='y = 0')
        plt.xlabel('Microchip Test 1')
        plt.ylabel('Microchip Test 2')
        plt.legend()
        plt.grid(True)
        plt.show()

data = np.loadtxt('/Users/Zohair/Library/Mobile Documents/com~apple~CloudDocs/Education/Machine Learning/machine-learning-ex2/ex2/ex2data2.txt', delimiter = ",")

# Run Logistic Regression

(m, n) = np.shape(data)

# Add Column Of Ones
temp = np.ones((m, 1)) 
data = np.hstack((temp, data))

(m, n) = np.shape(data)

# Get X and y from data
(X, y) = np.hsplit(data, (n - 1,))

# Feature Mapping
mappedX = mapFeature(X[:,1],X[:,2])

#Divide the sample into two: ones with positive classification, one with null classification
pos = np.array([X[i] for i in range(np.shape(X)[0]) if y[i] == 1])
neg = np.array([X[i] for i in range(np.shape(X)[0]) if y[i] == 0])

# Initialise Theta
initialTheta = np.zeros((mappedX.shape[1],1))

# Paramaters

lmbda = 1

# Initial Cost And Gradient
cost = computeCost(initialTheta, mappedX, y, lmbda)
grad = computeGradient(mappedX, y, initialTheta, lmbda)

print ("Cost at initial theta (zeros)", cost)

# Minimise Cost

# y must be 1-D for minimisation functions to work
y = y.flatten() 

def testFunction(theta):
     return computeCost(theta, mappedX, y, lmbda)

def testFunctionTwo(theta):
    return computeGradient(mappedX, y, theta, lmbda)

# thetaFinal = sp.fmin(computeCost, x0 = initialTheta, args = (mappedX, y, lmbda), maxiter=50000)
thetaFinal = sp.fmin_cg(testFunction, x0=initialTheta, fprime=testFunctionTwo)
# thetaFinal = sp.fmin(testFunction, x0 = initialTheta)

print(thetaFinal)

p = predict(thetaFinal, mappedX)
print ("Train Accuracy:", np.mean(p == y) * 100, "%")

# Plot Data
plotData()


```
